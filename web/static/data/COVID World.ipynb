```python
import pandas as pd
import json
```

```python
covid = "data.csv"
covid_df = pd.read_csv(covid)
covid_df.head()
```

```python
del covid_df['city']
covid_df
```

```python
covid_df.count()
```

```python
covid_world = covid_df[["county", "state", "country", "cases", "deaths", "lat", "long"]]
covid_world
```

```python
covid_world = covid_world.dropna(subset=['lat'])
```

```python
covid_world = covid_world.dropna(subset=['long'])
```

```python
covid_world = covid_world[covid_world['country'] != 'USA']
```

```python
covid_world = covid_world[covid_world['cases'] != 0]
```

```python
covid_world.count()
```

```python
covid_world.dtypes
```

```python
covid_world = covid_world.reset_index(drop=True)
```

```python
covid_world
```

```python
covid_world.to_json(r'covid_world.json', orient = 'records')
```
