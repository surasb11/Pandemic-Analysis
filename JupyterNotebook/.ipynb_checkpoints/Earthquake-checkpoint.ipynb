## DATA SHOWING AVERAGE LAND TEMPERATURES AND AVERAGE EARTHQUAKE MAGNITUDES FOR COUNTRIES TURKEY, IRAN, GREECE AND PAKISTAN DURING YEARS 2000 - 2013 

## LAND TEMPERATURE DATA

```python
# Import GlobalLandTemperatures_ByCountry.csv file
raw_Global_data = pd.read_csv(r'../csv-files/output_data/GlobalLandTemperatures.csv', error_bad_lines=False)
raw_Global_data.head()
```

```python
# YEARS 2000 - 2013
GlobalLand_data = raw_Global_data[(raw_Global_data['dt'] > '1999-12-31')]
GlobalLand_data['Year'] = pd.DatetimeIndex(GlobalLand_data['dt']).year
GlobalLand_data.head()
```

```python
GlobalLand_data.reset_index(inplace = True)
GlobalLand_data.head()
```

```python
GlobalLand_renamed = GlobalLand_data.rename(columns={"index": "index_la"})
GlobalLand_renamed.head()
```

```python
GlobalLand_clean = GlobalLand_renamed.drop(["AverageTemperatureUncertainty","index_la","dt"], axis=1)
GlobalLand_clean.head()
```

```python
GlobalLand_clean['Country'] = GlobalLand_clean['Country'].str.upper() 
GlobalLand_clean['Country'] = GlobalLand_clean['Country'].replace({'UNITED STATES': 'USA'})
Global_Temp = GlobalLand_clean.dropna(subset = ["Country", "AverageTemperature"])
Global_Temp = Global_Temp.set_index('Country')
GlobalLand_bycountry = Global_Temp.loc[["TURKEY", "IRAN","GREECE", "PAKISTAN"]]
GlobalLand_bycountry.head()
```

### CLEAN LAND TEMPERATURE DATAFRAME

```python
# MEAN LAND TEMPERATURE FOR COUNTRIES TURKEY, IRAN, GREECE, PAKISTAN
# GROUPED BY YEAR
Global_Temp_final = GlobalLand_bycountry.groupby(['Year','Country']).mean()
Global_final = pd.DataFrame(Global_Temp_final)
Global_final.head()
```

```python
def f(x):
    x = x * 1.8 + 32
    return float(x)
Global_final['AverageTemperature'] = Global_final['AverageTemperature'].apply(f)
Global_final.head()
```

## EARTHQUAKE DATA

```python
# Import earthquakes.csv file
raw_Earthquakes_data = pd.read_csv(r'../csv-files/output_data/Earthquakes.csv', error_bad_lines=False)
raw_Earthquakes_data.head()
```

```python
raw_Earthquakes_data['Year'] = pd.DatetimeIndex(raw_Earthquakes_data['date']).year
raw_Earthquakes_data.head()
```

```python
# YEARS 2000 - 2013
Earthquakes_data = raw_Earthquakes_data[(raw_Earthquakes_data['Year'] < 2014)]
Earthquakes_data.head()
```

```python
Earthquakes_data.reset_index(inplace = True)
Earthquakes_data.head()
```

```python
Earthquakes_renamed = Earthquakes_data.rename(columns={"index": "index_la", "mag": "AverageMagnitude", "country": "Country"})
Earthquakes_renamed.head()
```

```python
Earthquakes_clean = Earthquakes_renamed.drop(["index_la", "date", "latitude", "longitude",
                                               "depth", "net", "updated", "place", "depthError"], axis=1)
Earthquakes_clean.head()
```

```python
Earthquakes_clean['Country'] = Earthquakes_clean['Country'].str.upper() 
Earthquakes_clean['Country'] = Earthquakes_clean['Country'].replace({'US': 'USA'})
Earthquakes_clean.head()
```

```python
Earthquakes_Mag = Earthquakes_clean.dropna(subset = ["Country", "AverageMagnitude"])
Earthquakes_Mag.head()
```

```python
Earthquakes_Mag = Earthquakes_Mag.set_index('Country')
Earthquakes_Mag.head()
```

```python
Earthquakes_bycountry = Earthquakes_Mag.loc[["TURKEY", "IRAN","GREECE", "PAKISTAN"]]
Earthquakes_bycountry.head()
```

### CLEAN EARTHQUAKE MAGNITUDE DATAFRAME

```python
# MEAN EARTHQUAKE VALUES FOR COUNTRIES TURKEY, IRAN, GREECE, PAKISTAN
# GROUPED BY YEAR
Earthquakes_final = Earthquakes_bycountry.groupby(['Year','Country']).mean()
Earthquakes_final = pd.DataFrame(Earthquakes_final)
Earthquakes_final.head()
```

## FINAL DATAFRAME

```python
# MERGED DATAFRAME COUNTAINING AVERAGE LAND TEMPERATURES AND AVERAGE EARTHWUAKE MAGNITUDE VALUES PER YEAR
# FOR COUNTRIES AFGHANISTAN, TURKEY, IRAN, AZERBAIJAN, GREECE, PAKISTAN
# DURING YEARS 2000-2013
Earthquake_Temp = pd.merge(Earthquakes_final, Global_final, on=["Country", "Year"])
#Earthquake_Temp.reset_index(inplace=True)
Earthquake_Temp.head()
```
